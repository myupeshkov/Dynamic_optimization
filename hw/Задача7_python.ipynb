```python
from gekko import GEKKO
import numpy as np
import matplotlib.pyplot as plt
import math
```

# a)

```python
m = GEKKO(remote = False)
nt = 51
m.time = np.linspace(0,10,nt)

y = m.Var(value=0)
z = m.Var(value=1)
t = m.Var(value = 0)
u = m.Var(value=0)
p = np.zeros(nt)
p[-1] = 1.0
final = m.Param(value=p)

m.Equation(y.dt()== u)
m.Equation(t.dt() == 1)
m.Equation(z.dt()== (y-m.cos(t))**2 + (u ** 2))
m.Obj(z*final)
m.options.IMODE = 6
m.solve(disp=False)
plt.plot(m.time[1:], y.value[1:], 'k-', label=r'$y$')
plt.plot(m.time[1:],u.value[1:],'r--',label=r'$u$')
plt.legend(loc='best')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
print("Значение функционала: {}".format(m.options.OBJFCNVAL))
```
